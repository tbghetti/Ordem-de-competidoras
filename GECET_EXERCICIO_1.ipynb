```python
import numpy as np

atleta = list()

while True:
  nome_atleta = (str(input('Qual o nome da atleta? ')))

  n = 1;
  notas = np.zeros((n));

  for i in range(0, n):

    nota1 = float(input("Entre com a nota: "))
    print('Este é o placar da rodada 1: ', nota1)

    nota2 = float(input("Entre com a nota: "))
    soma1 = nota1 + nota2
    print('Este é o placar da rodada 2: ', soma1)

    nota3 = float(input("Entre com a nota: "))
    soma2 = nota1 + nota2 + nota3
    print('Este é o placar da rodada 3: ', soma2)

    nota4 = float(input("Entre com a nota: "))
    soma3 = nota1 + nota2 + nota3 + nota4
    print('Este é o placar da rodada 4: ', soma3)

    nota5 = float(input("Entre com a nota: "))
    notas1 = nota1, nota2, nota3, nota4, nota5

    notas_ordenadas1 = np.array(sorted(notas1))

    notas_mais_altas1 = notas_ordenadas1[-4:]

    soma_notas = np.sum(notas_mais_altas1)
    print('Este é o placar da rodada 5: ', soma_notas)

    nota6 = float(input("Entre com a nota: "))
    notas2 = nota1, nota2, nota3, nota4, nota5, nota6

    notas_ordenadas2 = np.array(sorted(notas2))

    notas_mais_altas2 = notas_ordenadas2[-4:]

    soma_notas2 = np.sum(notas_mais_altas2)
    print('Este é o placar da rodada 6: ', soma_notas2)

    nota7 = float(input("Entre com a nota: "))
    notas3 = nota1, nota2, nota3, nota4, nota5, nota6, nota7

    notas_ordenadas3 = np.array(sorted(notas3))

    notas_mais_altas3 = notas_ordenadas3[-4:]

    soma_notas3 = np.sum(notas_mais_altas3)

    print('Este é o placar final: ', soma_notas3)

  atleta.append([nome_atleta, soma_notas3])

  pergunta = str(input('Quer inserir mais uma atleta? [S/N] '))

  if pergunta in 'Nn':
    break

colocacao_final = sorted(atleta, key=lambda item: (100-item[1]))
print('Esta é a ordem da colocação final das atletas: ', colocacao_final)
```
